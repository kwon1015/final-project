```python
import pandas as pd
```

```python
# 1. Mecab 
from konlpy.tag import Mecab
from pprint import pprint
tokenizer= Mecab()


def get_VV_VA(stu_ans):
    pos = tokenizer.pos(stu_ans)
    pos = [p for p in pos if p[1] in ['VV', "VA"]]
    return pos
```

```python
val_v2 = pd.read_csv('./overthreshold_0.7.csv')
```

```python
val_v2
```

```python
val_v2['vv_va']= val_v2['sent_b'].apply(get_VV_VA)
```
