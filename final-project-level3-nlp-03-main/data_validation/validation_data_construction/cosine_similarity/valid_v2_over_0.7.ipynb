### 0.7 이상 pair만 건지기

```python
import os
```

```python
path = './'
questions = os.listdir(path)
questions = [q for q in questions if 'csv' in q]
```

```python
questions
```

```python
import pandas as pd

over_threshold = pd.DataFrame(columns = ['sent_a', 'sent_b', 'cosine_similarity', 'label'])

for q in questions:
    q_df = pd.read_csv(q)
    q_threshold = q_df[q_df['similarity'] > 0.7]
    label = [1 for _ in range(len(q_threshold))]
    q_threshold['label'] = label
    
    q_threshold = q_threshold.rename(columns={'right_ans':'sent_a'})
    q_threshold = q_threshold.rename(columns={'student_ans':'sent_b'})
    q_threshold = q_threshold.rename(columns={'similarity':'cosine_similarity'})
    
    q_threshold.to_csv(f'./validation_v2/{q.split(".")[0]}_over_0.7.csv')
```

```python
over_threshold.drop(['Unnamed: 0'], axis=1, inplace=True)
```
