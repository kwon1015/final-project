## 질문 별 키워드 분포 묶음 만들기

```python
q_51_pd.loc[:]
```

```python
for wai_q in wai_questions :
    q_pd = pd.read_csv(os.path.join(wai_path, wai_q))
    cols = q_pd.columns
    
    # remove None 
    q_col = cols[1]
    print(q_col)
    q_pd = q_pd[q_pd[q_col]==q_pd[q_col]] 
    
    # remove duplicate
    q_pd = q_pd.drop_duplicates([q_col])

    
    # get keyword list
    keywords = [col.split("_")[1] for col in cols if (col[0]=="q" and col[-1]!="1")]
    print(keywords)
    
    
    # keyword matching func
    def keywords_matching(answer):
        #idx = []
        num = 0
        for kwd in keywords:
            if kwd in answer:
                #idx.append(answer.index(kwd))
                num+=1
        return num
    
    # count keywords
    q_pd["kwd_match_num"] = q_pd[q_col].apply(keywords_matching)
    
    # group answers by the number of matching keywords
    matching_groups = dict(q_pd["kwd_match_num"].value_counts())
    print(matching_groups)
    
    keywords.sort(reverse=True, key=lambda x : len(x))
    def add_keyword_tokens(answer):
        answer_with_kwd = answer
        for kwd in keywords:
            if kwd in answer:
                idx = answer_with_kwd.index(kwd)
                length = len(kwd)
                answer_with_kwd = answer_with_kwd[:idx] + "[kwd]" + answer_with_kwd[idx:idx+length] + "[kwd]" + answer_with_kwd[idx+length:]       
        return answer_with_kwd
                    
    for k, v in matching_groups.items():
        kwd_pd = q_pd[q_pd["kwd_match_num"]==k]
        kwd_pd.iloc[:,1] = kwd_pd[q_col].apply(add_keyword_tokens)
        kwd_pd = kwd_pd.drop([col for col in cols if "Unnamed" in col], axis = 1)
        kwd_pd.to_csv(f'wai_keywords/kwd_num_{k}/{wai_q}-keyword_match_num_{k}.csv')
```
